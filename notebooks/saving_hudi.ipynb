```python
import pyspark
import pyspark.sql.functions as F
import os
import pandas as pd
```

```python

cwd = os.getcwd()
parent_dir = os.path.dirname(cwd)
os.chdir(parent_dir)
```

```python
from pyspark.sql import SparkSession
from delta import configure_spark_with_delta_pip

builder = (
    SparkSession.builder 
    .master("local[8]")
    .appName("hudi")
    .config("spark.driver.memory", "13g")
    .config("spark.jars", "/home/eric-yoshida/.ivy2/jars/hudi-spark3.5-bundle_2.12-1.0.2.jar")
    .config("spark.serializer", "org.apache.spark.serializer.KryoSerializer")
)

spark = configure_spark_with_delta_pip(builder).getOrCreate()
```

```python
data_path = "./data/processed/"

united_df = spark.read.json(data_path + "united_df.json")
```

```python
united_df.limit(10).toPandas()
```

```python
united_df.printSchema()
```

```python
hudi_path = os.getcwd() + "/data/hudi/united_data"

(
    united_df.write
        .format("hudi")
        .option("hoodie.table.name", "united_data")
        .option("hoodie.datasource.write.table.type", "MERGE_ON_READ")
        .option("hoodie.base.file.format", "ORC")
        .option("hoodie.datasource.write.recordkey.field", "reviewerID")
        .option("hoodie.datasource.write.precombine.field", "unixReviewTime")
        .mode("overwrite")
        .option("overwriteSchema", "True")
        .save(hudi_path)
 )
```
